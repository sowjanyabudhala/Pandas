```python
# Convert object to category
category_list = ['name', 'full-name', 'alter-egos', 'aliases', 'place-of-birth', 'first-appearance',
                  'publisher', 'alignment', 'gender', 'race', 'eye-color', 'hair-color',
                  'occupation', 'base', 'group-affiliation', 'relatives', 'url']

for i in df1.columns:
  if i in category_list:
    df1[i]=df1[i].astype('category')
```

```python
if 'height' in df1.columns:
    df1['height']=df1['height'].str.extract(r"(\d+'?\d*)")
    df1['height']=df1['height'].str.replace("'",'.').astype(float)
else:
  df1['height']=np.nan
```

```python
if 'weight' in df1.columns:
    df1['weight']=df1['weight'].str.extract(r"(\d+.\d+)")[0].astype(float)
else:
      df1['weight']=np.nan
```

**REMOVING/ADDING COLUMNS USING LAMBDA & CUSTOM FUNCTIONS**

```python
df1['height_category']=df1['height'].apply(lambda x:'short' if x<8.5 else('average' if x<5.0 else 'tall'))
```

```python
def categorize_heroes(row):
  if row['height']<2.7 and row['weight']<2:
    return 'light weight'
  elif row['height']<5.6 and row['weight']<4.6:
    return 'middle weight'
  else:
      return 'heavy weight'
```

```python
df1['category']=df1.apply(categorize_heroes,axis=1)
```
